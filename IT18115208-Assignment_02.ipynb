5)	Consider the url=’http://www.dictionary.com/browse/’ and the words ‘handy’,’whisper’,’lovely’,’scrape’.
Build a data frame, where the first variable is “Word” and the second variable is “definitions”. Scrape the definitions from the url.


```python
import pandas as pd
from urllib.request import urlopen
from bs4 import BeautifulSoup

#dictionary with the column names of the data frame
Dictionary_df = { 'Word':[] , 'Definitions':[] };

#List of the words
words = ['handy','whisper','lovely','scrape']    

for word in words:
    Dictionary_df['Word'] += [word]    
    url = "https://www.dictionary.com/browse/" + word + "?s=t"
    html = urlopen(url)
    
    soup = BeautifulSoup(html, 'html.parser')

    definition = soup.find('span',attrs={'class':'one-click-content css-1p89gle e1q3nk1v4'})

    Dictionary_df['Definitions'] += [definition.get_text()]

      
dataFrame = pd.DataFrame(Dictionary_df,columns=['Word','Definitions'])
dataFrame
```
